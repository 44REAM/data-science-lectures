## Machine learning vs Deep learning

Machine learning ยังต้องมี domain knowledge ช่วยอยู่ในการเลือกสิ่งที่เราจะส่งเข้าไปให้ model learn ซึ่งในบางครั้ง knowledge ของมนุษย์อาจจะยังมีไม่พอในการทำให้เครื่องสามารถเรียนได้อย่างมีประสิทธิภาพ


### Example

-   Face detection (unstructural data):
    -   ML: มนุษย์จะต้องทำ features extraction เช่นจากรูปหน้าคน เราอาจจะ extract ออกมาว่าตาขนาดเท่าไร ตาห่างกับจมูกเท่าไร แล้วนำค่าพวกนี้ใส่ให้ model learn
    -   DL: สามารถใส่รูปไปเพื่อให้เครื่องเรียนรู้ได้ด้วยตัวเองเลย

-   tabular data เช่นข้อมูลผู้ป่วย เพศ อายุ etc (structural data):
    -   ML: เราอาจจะไม่จำเป็นต้องทำ features extraction เลยก็ได้
    -   DL: ทำได้เหมือนกันกับ ML แต่ตัว model จะ complex มากกว่า

จะเห็นว่าถ้าข้อมูลที่เราต้องการให้เครื่องเรียนรู้มีความซับซ้อนมากขึ้นเช่น รูปภาพหรือข้อมูลที่อยู่ในรูปแบบข้อความ (หนังสือ) DL นั้นมีความสำคัญอย่างมาก

# Type of machine learning

- Supervised learning: เรียนจากข้อมูลที่มีอยู่และมีคำตอบแล้ว จะพยายามหาความสัมพันธ์ระหว่าง input features กับ labels (target) (เฉลยคำตอบ)
- Unsupervised learning: เรียนจากข้อมูลที่ไม่ได้มีคำตอบ เรียนเพื่อดู pattern ข้อมูล
- Reinforcement learning: self-learning เรียนจาก response (feedback) จาก environment reinforcement learning จะต่างกับ supervised learning ตรงที่ supervised learning จะพยายามเรียนจากสิ่งที่เรา input เข้าไป แต่ reinforcement learning จะสามารถ explore option และหาคำตอบที่ถูกได้ด้วยตัวเองจากการได้รับ reward

## Learning algorithm example

Learning algorithm คือวิธีการที่ model ใช้ในการ optimized parameter ใน model, $θ$ คือ parameter ของ model เช่นถ้าเป็น linear model $θ$ คือ $a_0, a_1$ error หรือเรียกอีกชื่อว่า loss คือ error ของค่าที่ predicted จาก model เราว่ามีค่าต่างจากค่าจริงๆเท่าไร (หา parameter ที่ทำให้ error น้อยที่สุด)

ตัวอย่าง learning algorithm เช่น gradient descent ดังรูปข้างล่าง จะพยายามปรับค่า $θ$ เพื่อให้ได้ error ที่น้อยที่สุด


***Learning rate*** ใน gradient descent ใช้เพื่อปรับความเร็วในการเรียนของ algorithm

***hyperparameter*** คือ parameter ที่ใช้ใน learning process เช่น learning rate ถือเป็น hyperparameter

## Linear regression

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 14, 18, 25, 35, 38]).reshape((-1, 1))
```

```python
plt.scatter(x, y)
```

$$y = a_0 + a_1 x$$

```python
# เลือกว่าจะใช้ model linear regression
model = LinearRegression()

# (training) สั่ง model เรียนจากข้อมูล
model.fit(x, y)

# (test on training set) เมื่อได้ model ที่เรียนแล้วเรานำ model ที่เรียนมาลองใช้งานกับข้อมูล
pred = model.predict(x)
```

```python
y
```

```python
pred
```

$$MSE = \frac{1}{n} ∑ (y - pred)^2 $$

```python
from sklearn.metrics import mean_squared_error
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html


mean_squared_error(y, pred)
```

$$y = 2.36 + 0.67 x$$

```python
print(f"intercept: {model.intercept_}")


print(f"slope: {model.coef_}")
```

## Linear regression with train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

print(f"x.shape {x.shape}")
print(f"X_train.shape {X_train.shape}")
print(f"X_test.shape {X_test.shape}")
```

```python
model = LinearRegression()

# training
model.fit(X_train, y_train)

# test on training set
train_pred = model.predict(X_train)

# test on test set
test_pred = model.predict(X_test)
```

```python
test_pred
```

```python
y_test
```

```python
print(f"test MSE {mean_squared_error(y_test, test_pred)}")
print(f"train MSE {mean_squared_error(y_train, train_pred)}")
```

```python
from sklearn.datasets import load_iris#save data information as variable
iris = load_iris()
```

features ในที่นี้ของเราคือความกว้างและความยาวของกลีบใบ (Sepal and Petal)

```python
print(iris['DESCR'])
```

```python
np.unique(iris['target'], return_counts = True)
```

```python
# ทำให้เป็น imbalance dataset

x_1 = iris['data'][(iris['target'] == 1)]
x_2 = iris['data'][(iris['target'] == 2)][:5]

# เอาข้อมูลมาต่อกัน
x = np.concatenate([x_1, x_2])

y_1 = iris['target'][(iris['target'] == 1)]
y_2 = iris['target'][(iris['target'] == 2)][:5]

# เอาข้อมูลมาต่อกัน
y = np.concatenate([y_1, y_2])
```

```python
np.unique(y, return_counts = True)
```

```python
iris['data'].shape
```

```python
x.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=42, stratify = y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()

# training
model.fit(X_train, y_train)

# test on training set
train_pred = model.predict(X_train)

# test on test set
test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(f"train ACC {accuracy_score(y_train, train_pred)}")
print(f"test ACC {accuracy_score(y_test, test_pred)}")
```

```python
y_test
```

```python
test_pred
```

```python
from sklearn.metrics import balanced_accuracy_score

print(f"train BACC {balanced_accuracy_score(y_train, train_pred)}")
print(f"test BACC {balanced_accuracy_score(y_test, test_pred)}")
```
